# PyCitySchools 

```python
# Dependencies and Setup
import pandas as pd
import numpy as np
```

```python
# File to Load 
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"
```

```python
# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load, encoding="utf-8")
school_data.head()
```

```python
student_data = pd.read_csv(student_data_to_load, encoding="utf-8")
student_data.head()
```

```python
# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# View column names
school_data_complete.columns
```

```python
#Calculate the total number of schools

total_school = len(school_data_complete['school_name'].unique())

total_school
```

```python
#Calculate the total number of students
total_student_no = len(school_data_complete['Student ID'].unique())
total_student_no
```

```python
#Calculate the total budget
total_budget = school_data['budget'].sum()
total_budget
```

```python
#Calculate the average math score
avg_math_score = school_data_complete['math_score'].mean()
avg_math_score
```

```python
#Calculate the average reading score
avg_read_score = school_data_complete['reading_score'].mean()
avg_read_score
```

```python
#Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2
passing_rate = (avg_math_score + avg_read_score)/2
passing_rate
```

```python
# Calculate the percentage of students with a passing math score (70 or greater)
pass_math = (school_data_complete.loc[school_data_complete['math_score']>=70])
pass_math_percent = (len(pass_math['math_score'])/total_student_no) * 100
pass_math_percent
```

```python
pass_math.head()
```

```python
# Calculate the percentage of students with a passing reading score (70 or greater)
pass_read = (school_data_complete.loc[school_data_complete['reading_score']>=70])
pass_read_percent = ((len(pass_read['reading_score'])/total_student_no) * 100)
pass_read_percent
```

```python
pass_read.head()
```

```python
# Create a dataframe to hold the above results
district_df = pd.DataFrame({
    "Total Schools": [total_school],
    "Total Students": [total_student_no], 
    "Total Budget": [total_budget],
    "Average Math Score": [avg_math_score],
    "Average Reading Score": [avg_read_score],
    "Overall Passing Rate": [passing_rate],
    "Passing Math Score %": [pass_math_percent],
    "Passing Reading Score %": [pass_read_percent]
})

# Optional: give the displayed data cleaner formatting

district_df["Total Budget"] = district_df["Total Budget"].map("${:,.2f}".format)
district_df["Average Math Score"] = district_df["Average Math Score"].map("{:.2f}".format)
district_df["Average Reading Score"] = district_df["Average Reading Score"].map("{:.2f}".format)
district_df["Overall Passing Rate"] = district_df["Overall Passing Rate"].map("{:.2f}".format)
district_df["Passing Math Score %"] = district_df["Passing Math Score %"].map("{:.2f}".format)
district_df["Passing Reading Score %"] = district_df["Passing Reading Score %"].map("{:.2f}".format)
district_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
school_summary = school_data_complete.groupby(['school_name'])
sch_type = school_summary['type'].first()
sch_students = school_summary['Student ID'].count()
sch_budget = school_summary['budget'].first()
per_student_budget = sch_budget/sch_students
sch_avg_math = school_summary['math_score'].mean()
sch_avg_read = school_summary['reading_score'].mean()
```

```python
sch_pass_math = pass_math.groupby(['school_name'])
sch_pass_math_percent = (sch_pass_math['math_score'].count()/sch_students) * 100
sch_pass_read = pass_read.groupby(['school_name'])
sch_pass_read_percent = (sch_pass_read['reading_score'].count()/sch_students) * 100
sch_overall = (sch_pass_math_percent + sch_pass_read_percent)/2
```

```python
#unformatted
sch_summary_0 = pd.DataFrame({
    "School Type": sch_type, 
    "Total Students": sch_students,
    "Total School Budget": sch_budget,
    "Per Student Budget": per_student_budget,
    "Average Math Score": sch_avg_math,
    "Average Reading Score": sch_avg_read,
    "% Passing Math": sch_pass_math_percent,
    "% Passing Reading":sch_pass_read_percent,
    "Overall Passing Rate": sch_overall
})
```

```python
sch_summary_df = pd.DataFrame({
    "School Type": sch_type, 
    "Total Students": sch_students,
    "Total School Budget": sch_budget,
    "Per Student Budget": per_student_budget,
    "Average Math Score": sch_avg_math,
    "Average Reading Score": sch_avg_read,
    "% Passing Math": sch_pass_math_percent,
    "% Passing Reading":sch_pass_read_percent,
    "Overall Passing Rate": sch_overall
})

sch_summary_df["Total Students"] = sch_summary_df["Total Students"].map("{:,}".format)
sch_summary_df["Total School Budget"] = sch_summary_df["Total School Budget"].map("${:,.2f}".format)
sch_summary_df["Per Student Budget"] = sch_summary_df["Per Student Budget"].map("${:,.2f}".format)
sch_summary_df["Average Math Score"] = sch_summary_df["Average Math Score"].map("{:.2f}".format)
sch_summary_df["Average Reading Score"] = sch_summary_df["Average Reading Score"].map("{:.2f}".format)
sch_summary_df["% Passing Math"] = sch_summary_df["% Passing Math"].map("{:.2f}%".format)
sch_summary_df["% Passing Reading"] = sch_summary_df["% Passing Reading"].map("{:.2f}%".format)
sch_summary_df["Overall Passing Rate"] = sch_summary_df["Overall Passing Rate"].map("{:.2f}%".format)
sch_summary_df
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
# "Average Math Score" for each grade level (9th, 10th, 11th, 12th) at each school

math_table = school_data_complete[["school_name", "grade", "math_score"]]
# grade 9
math_9 = math_table.loc[math_table["grade"] == "9th"].groupby("school_name")
avg_math_9 = math_9["math_score"].mean()

# grade 10
math_10 = math_table.loc[math_table["grade"] == "10th"].groupby("school_name")
avg_math_10 = math_10["math_score"].mean()

# grade 11
math_11 = math_table.loc[math_table["grade"] == "11th"].groupby("school_name")
avg_math_11 = math_11["math_score"].mean()

#grade 12
math_12 = math_table.loc[math_table["grade"] == "12th"].groupby("school_name")
avg_math_12 = math_12["math_score"].mean()


math_scores = pd.DataFrame({
    "9th":avg_math_9,
    "10th":avg_math_10,
    "11th":avg_math_11,
    "12th":avg_math_12
})

math_scores["9th"] = math_scores["9th"].map("{:.2f}".format)
math_scores["10th"] = math_scores["10th"].map("{:.2f}".format)
math_scores["11th"] = math_scores["11th"].map("{:.2f}".format)
math_scores["12th"] = math_scores["12th"].map("{:.2f}".format)
math_scores
```

## Reading Score by Grade 

* Perform the same operations as above for reading scores

```python
# "Average Reading Score" for each grade level (9th, 10th, 11th, 12th) at each school

read_table = school_data_complete[["school_name", "grade", "reading_score"]]
# grade 9
read_9 = read_table.loc[math_table["grade"] == "9th"].groupby("school_name")
avg_read_9 = read_9["reading_score"].mean()

# grade 10
read_10 = read_table.loc[math_table["grade"] == "10th"].groupby("school_name")
avg_read_10 = read_10["reading_score"].mean()

# grade 11
read_11 = read_table.loc[math_table["grade"] == "11th"].groupby("school_name")
avg_read_11 = read_11["reading_score"].mean()

#grade 12
read_12 = read_table.loc[math_table["grade"] == "12th"].groupby("school_name")
avg_read_12 = read_12["reading_score"].mean()


read_scores = pd.DataFrame({
    "9th":avg_read_9,
    "10th":avg_read_10,
    "11th":avg_read_11,
    "12th":avg_read_12
})
read_scores["9th"] = read_scores["9th"].map("{:.2f}".format)
read_scores["10th"] = read_scores["10th"].map("{:.2f}".format)
read_scores["11th"] = read_scores["11th"].map("{:.2f}".format)
read_scores["12th"] = read_scores["12th"].map("{:.2f}".format)
read_scores
```

  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
